## Faster R-CNN R50

```python
# Chuẩn bị config

from mmdet.apis import set_random_seed
from mmcv import Config
cfg = Config.fromfile('/content/drive/MyDrive/CS114.M11/mmdetection/configs/faster_rcnn/faster_rcnn_r50_fpn_1x_coco.py')

cfg.gpu_ids = [0]
# Modify dataset type and path
cfg.data_root = data_root_dir

cfg.data.test.type = 'UITVinaDeveS22'
cfg.data
cfg.data.test.data_root = '{}/'.format(data_root_dir) 
cfg.data.test.ann_file = 'outputtest.json'
cfg.data.test.img_prefix = 'images'

cfg.data.train.type = 'UITVinaDeveS22'
cfg.data.train.data_root = '{}/'.format(data_root_dir) 
cfg.data.train.ann_file = 'outputtrain.json'
cfg.data.train.img_prefix = 'images'

cfg.data.val.type = 'UITVinaDeveS22'
cfg.data.val.data_root = '{}/'.format(data_root_dir) 
cfg.data.val.ann_file = 'outputvalid.json'
cfg.data.val.img_prefix = 'images'

# Số class
cfg.model.roi_head.bbox_head.num_classes = 8

# Đường dẫn lưu checkpoints
cfg.work_dir = '/content/drive/MyDrive/CS114.M11/work_dirs/faster_rcnn_r50_fpn_1x_coco'

# The original learning rate (LR) is set for 8-GPU training.
# We divide it by 8 since we only use one GPU.
cfg.optimizer.lr = 0.02 / 8
cfg.lr_config.warmup = None
cfg.log_config.interval = 500

# Change the evaluation metric since we use customized dataset.
#cfg.evaluation.metric = 'mAP'
# We can set the evaluation interval to reduce the evaluation times
cfg.evaluation.interval = 1
# We can set the checkpoint saving interval to reduce the storage cost
cfg.checkpoint_config.interval = 3

# Set seed thus the results are more reproducible
cfg.seed = 0
set_random_seed(0, deterministic=False)
cfg.gpu_ids = range(1)
cfg.device = 'cuda'

cfg.dump(os.path.join(cfg.work_dir, os.path.basename('faster_rcnn_r50_fpn_1x_coco.py')))
```

```python
from mmdet.datasets import build_dataset
from mmdet.models import build_detector
from mmdet.apis import train_detector
import os

# Build dataset
datasets = [build_dataset(cfg.data.train)]

# Build the detector
model = build_detector(cfg.model, train_cfg=cfg.get('train_cfg'), test_cfg=cfg.get('test_cfg'))
# Add an attribute for visualization convenience
model.CLASSES = ('bicycle', 'motorcycle', 'car', 'van', 'truck', 'bus', 'fire truck',  'container')

# Create work_dir
if not os.path.exists(cfg.work_dir):
    os.mkdir(cfg.work_dir)
```

```python
print(f'Config:\n{cfg.pretty_text}')
```

```python
# Training
train_detector(model, datasets, cfg, distributed=False, validate=True)
```
